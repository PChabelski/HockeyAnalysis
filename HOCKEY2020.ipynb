Feature Engineering

What kind of features can we develop from the existing data?

For a truculence analysis:

- Player "Density": Mass of the player / height of the player; Get an idea of the size of the player (tall and lanky, short and stocky)
- Battles: Hits + Hits Taken; indication of a players interaction with opposing players ---> Can do Battles/60, BattlesPlusMinus
- BU: Body Usage: Hits + Hits Taken + Shots Blocked; indication of a player's level of physicality (putting their body on the line) ---> can add BU/60, 
- WPS: Weighted Penalty Score: 10% Minor, 65% Major, 35% Misconduct; players with higher weighted penalty scores will (ideally) be fighters, dirtier, more truculent (filters out players that have a high number of minor stick infractions, for instance). --> These weights can be modified. 

```python
leafs_df['Density'] = leafs_df['Weight (lbs)']/leafs_df['Height (in)']
```

```python
leafs_df['BU'] = leafs_df['Hits']+leafs_df['Hits Taken']+leafs_df['Shots Blocked']
```

```python
leafs_df['BU/60'] = (leafs_df['BU']/leafs_df['TOI'])*60
```

```python
leafs_df['WPS'] = (0.1*leafs_df['Minor'])+(0.6*leafs_df['Major'])+(0.3*leafs_df['Misconduct'])
```

```python
leafs_df['WPS/60'] = (leafs_df['WPS']/leafs_df['TOI'])*60
```

```python
leafs_df['Battles'] = leafs_df['Hits']+leafs_df['Hits Taken']
```

```python
leafs_df['Battles/60'] = (leafs_df['Battles']/leafs_df['TOI'])*60
```

```python
leafs_df['BattlesPlusMinus'] = leafs_df['Hits']-leafs_df['Hits Taken']
```

```python
leafs_df['Major/60'] =(leafs_df['Major']/leafs_df['TOI'])*60    # Find the Scrappiest
```

```python
leafs_df['Minor/60'] =(leafs_df['Minor']/leafs_df['TOI'])*60    # Find the most undisciplined
```

```python
leafs_df['Misconduct/60'] =(leafs_df['Misconduct']/leafs_df['TOI'])*60    # Find the most insane
```
